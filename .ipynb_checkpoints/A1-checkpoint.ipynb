## Step1

```python
import json
import requests
```

```python
endpoint_legacy = 'https://wikimedia.org/api/rest_v1/metrics/legacy/pagecounts/aggregate/{project}/{access-site}/{granularity}/{start}/{end}'

endpoint_pageviews = 'https://wikimedia.org/api/rest_v1/metrics/pageviews/aggregate/{project}/{access}/{agent}/{granularity}/{start}/{end}'
```

```python
# SAMPLE parameters for getting aggregated legacy view data 
# see: https://wikimedia.org/api/rest_v1/#!/Legacy_data/get_metrics_legacy_pagecounts_aggregate_project_access_site_granularity_start_end
desktop_params_legacy = {"project" : "en.wikipedia.org",
                 "access-site" : "desktop-site",
                 "granularity" : "monthly",
                 "start" : "2007120100",
                # for end use 1st day of month following final month of data
                 "end" : "2016080100"
                    }

mobile_params_legacy = {"project" : "en.wikipedia.org",
                 "access-site" : "mobile-site",
                 "granularity" : "monthly",
                 "start" : "2007120100",
                # for end use 1st day of month following final month of data
                 "end" : "2016080100"
                    }

# SAMPLE parameters for getting aggregated current standard pageview data
# see: https://wikimedia.org/api/rest_v1/#!/Pageviews_data/get_metrics_pageviews_aggregate_project_access_agent_granularity_start_end
desktop_params_pageviews = {"project" : "en.wikipedia.org",
                    "access" : "desktop",
                    "agent" : "user",
                    "granularity" : "monthly",
                    "start" : "2015070100",
                    # for end use 1st day of month following final month of data
                    "end" : '2021101000'
                        }

mobile_web_params_pageviews = {"project" : "en.wikipedia.org",
                    "access" : "mobile-web",
                    "agent" : "user",
                    "granularity" : "monthly",
                    "start" : "2015070100",
                    # for end use 1st day of month following final month of data
                    "end" : '2021101000'
                        }

mobile_app_params_pageviews = {"project" : "en.wikipedia.org",
                    "access" : "mobile-app",
                    "agent" : "user",
                    "granularity" : "monthly",
                    "start" : "2015070100",
                    # for end use 1st day of month following final month of data
                    "end" : '2021101000'
                        }

# Customize these with your own information
headers = {
    'User-Agent': 'https://github.com/zhangjunhao0',
    'From': 'zjh01@uw.edu'
}
```

```python
def api_call(endpoint,parameters):
    call = requests.get(endpoint.format(**parameters), headers=headers)
    response = call.json()
    
    return response
```

```python
desktop_monthly_pageviews = api_call(endpoint_pageviews, desktop_params_pageviews)
mobile_web_monthly_pageviews = api_call(endpoint_pageviews, mobile_web_params_pageviews)
mobile_app_monthly_pageviews = api_call(endpoint_pageviews, mobile_app_params_pageviews)
desktop_monthly_legacy = api_call(endpoint_legacy, desktop_params_legacy)
mobile_monthly_legacy = api_call(endpoint_legacy, mobile_params_legacy)
```

```python
print(desktop_monthly_pageviews)
```

```python
import json
def save_data_to_json(filename, data):
    with open(filename, 'w') as outfile:
        json.dump(data, outfile)
```

```python
save_data_to_json('pagecounts_desktop-site_200712-201608.json', desktop_monthly_legacy)
save_data_to_json('pagecounts_mobile-site_200712-201608.json', mobile_monthly_legacy)
save_data_to_json('pageviews_desktop-site_201507-202109.json', desktop_monthly_pageviews)
save_data_to_json('pageviews_mobile-web-site_201507-202109.json', mobile_web_monthly_pageviews)
save_data_to_json('pageviews_mobile-app-site_201507-202109.json', mobile_app_monthly_pageviews)
```

## Step 2

```python
def read_json(filename):
    with open(filename) as file:
        return json.load(file)
```

```python
pagecounts_desktop = read_json('pagecounts_desktop-site_200712-201608.json')
pagecounts_mobile = read_json('pagecounts_mobile-site_200712-201608.json')
pageviews_desktop = read_json('pageviews_desktop-site_201507-202109.json')
pageviews_mobile_web = read_json('pageviews_mobile-web-site_201507-202109.json')
pageviews_mobile_app = read_json('pageviews_mobile-app-site_201507-202109.json')
```

```python
# store into list of [timestamp, count]
pagecounts_desktop_data = []
for item in pagecounts_desktop['items']:
    pagecounts_desktop_data.append([item['timestamp'], item['count']])
pagecounts_mobile_data = []
for item in pagecounts_mobile['items']:
    pagecounts_mobile_data.append([item['timestamp'], item['count']])
pageviews_desktop_data = []
for item in pageviews_desktop['items']:
    pageviews_desktop_data.append([item['timestamp'], item['views']])
pageviews_mobile_web_data = []
for item in pageviews_mobile_web['items']:
    pageviews_mobile_web_data.append([item['timestamp'], item['views']])
pageviews_mobile_app_data = []
for item in pageviews_mobile_app['items']:
    pageviews_mobile_app_data.append([item['timestamp'], item['views']])
```

```python
import pandas as pd
```

```python
df_pagecounts_desktop = pd.DataFrame(pagecounts_desktop_data)
df_pagecounts_mobile = pd.DataFrame(pagecounts_mobile_data)
df_pageviews_desktop = pd.DataFrame(pageviews_desktop_data)
df_pageviews_mobile_web = pd.DataFrame(pageviews_mobile_web_data)
df_pageviews_mobile_app = pd.DataFrame(pageviews_mobile_app_data)
```

```python
df_pagecounts = df_pagecounts_desktop.merge(df_pagecounts_mobile, how='outer', on=0).fillna(0)
df_pagecounts = df_pagecounts.rename(columns={'1_x':'pagecount_desktop_views', '1_y':'pagecount_mobile_views'})
df_pagecounts['pagecount_all_views'] = df_pagecounts['pagecount_desktop_views']+df_pagecounts['pagecount_mobile_views']
```

```python
print(df_pagecounts)
```

```python
df_pageviews = df_pageviews_desktop.merge(df_pageviews_mobile_web, how='outer', on=0).merge(df_pageviews_mobile_app, how='outer', on=0).fillna(0)
df_pageviews = df_pageviews.rename(columns={'1_x':'pageview_desktop_views', '1_y':'pageview_mobile_web_views', 1:'pageview_mobile_app_views'})
df_pageviews['pageview_mobile_views'] = df_pageviews['pageview_mobile_web_views']+df_pageviews['pageview_mobile_app_views']
df_pageviews['pageview_all_views'] = df_pageviews['pageview_desktop_views']+df_pageviews['pageview_mobile_views']
```

```python
print(df_pageviews)
```

```python
df_all =  df_pagecounts.merge(df_pageviews, how='outer', on=0).fillna(0)
```

```python
df_all['year'] = df_all[0].apply(lambda x: x[:4])
df_all['month'] = df_all[0].apply(lambda x: x[4:6])
```

```python
print(df_all)
```

```python
df_final = df_all[['year','month','pagecount_all_views','pagecount_desktop_views','pagecount_mobile_views','pageview_all_views','pageview_desktop_views','pageview_mobile_views']]
```

```python
df_final.to_csv('en-wikipedia_traffic_200712-202108.csv')
```
